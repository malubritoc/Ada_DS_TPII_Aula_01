#### Ex. 1: Reproduza o gráfico de barras abaixo, em que cada barra representa um dos 10 países com mais casos confirmados de COVID no dataset, e a "quebra" em cada cor indica a predominância de casos confirmados em cada mês.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Covid_19_Countrywise_timeseries.csv")
df.head()
```

```python
paises_maiscasos = df.groupby("country").agg({"Confirmed": "max"}).nlargest(10, "Confirmed").index.tolist()
df = df[df["country"].isin(paises_maiscasos)]
```

```python
df['anomes'] = pd.to_datetime(df['ObservationDate']).astype('string').str.slice(0, 7, 1)
df

matriz_agregacao = pd.pivot_table(df, values='New Confirmed', index='country', columns='anomes', aggfunc='sum').fillna(0)

matriz_agregacao.plot(kind = 'barh', stacked=True)
```

#### Ex.2: Reproduza o gráfico de linhas abaixo, que representa a série temporal da evolução de emissões de CO2 no Brasil.


```python
df = pd.read_csv("co2_emissions_kt_by_country.csv")
df.head()
```

```python
df_brazil = df[df['country_name'] == 'Brazil']
df_brazil
```

```python
df_brazil.plot(kind = 'line', x = 'year', y = 'value', legend = False,
              title = 'Histórico de emissões de CO2 no Brasil',
              xlabel = 'Ano')
```

#### Ex. 3:  Uma visualização muito similar à anterior, mas com a adição da média global de emissão de CO2 (o Brasil deve ser incluído na linha da média global?).

```python
df_global = df.groupby('year')['value'].mean()
df_global
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(df_brazil['year'], df_brazil['value'], label='Brasil')
ax.plot(df_global, label='Média Global')
```

#### Ex.4: Os 10 países que mais aumentaram, percentualmente, a emissão de CO2 entre os anos de 2018 e 2019, na forma de um gráfico de barras horizontal.


```python
df_filtroano = df[(df['year'] == 2018) | (df['year'] == 2019)]
```

```python
df_agrupado = df_filtroano.groupby('country_name').agg({'value': lambda x: (x.iloc[1] - x.iloc[0]) / x.iloc[0] * 100})
df_ordem = df_agrupado.sort_values(by='value', ascending=False)
df_grafico = df_ordem.head(10)
df_grafico
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.barh(df_grafico.index, df_grafico['value'], color='green')

ax.set_xlabel('Variação percentual na emissão de CO2 entre 2018 e 2019')
ax.set_ylabel('Países')

plt.show()
```
